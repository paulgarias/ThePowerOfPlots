```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
df_clinical_data  = pd.read_csv('raw_data/clinicaltrial_data.csv')
df_mouse_drug_data = pd.read_csv('raw_data/mouse_drug_data.csv')
df_full_data = df_mouse_drug_data.set_index("Mouse ID").join(df_clinical_data.set_index("Mouse ID"))
df_full_data = df_full_data.reset_index()
```

```python
df_full_data.head()
```

```python
tumor_volume_over_time_per_treatment = df_full_data.groupby(["Drug","Timepoint"]).mean()["Tumor Volume (mm3)"].unstack(level=0)
tumor_volume_over_time_per_treatment.head(10)
```

```python
metastatic_sites_over_time_per_treatment = df_full_data.groupby(["Drug","Timepoint"]).mean()["Metastatic Sites"].unstack(level=0)
metastatic_sites_over_time_per_treatment.head()
```

```python
mouse_survival_over_time_per_treatment = df_full_data.groupby(["Drug","Timepoint"]).count()["Mouse ID"].unstack(level=0)
mouse_survival_over_time_per_treatment.head()
```

```python
percent_change_volume = ((tumor_volume_over_time_per_treatment.loc[45]-
                         tumor_volume_over_time_per_treatment.loc[0])
                         /tumor_volume_over_time_per_treatment.loc[0])*100
```
